### Question: Take an input from the user for start and end values. Print all even numbers between them.


```python
start = int(input())
end = int(input())


increment_count = 0
## start= 1 , end = 10
while start <=end:
    if (start)%2==0:
        print(start)
    start+=1
    increment_count += 1

print("Total Increments", increment_count)
```

### For Loop

```python
for i in range(1,5):
    print(i)
```

```python
for i in range(1,5,2):
    print("Hello")
```

```python
list(range(1,5,2))
```

```python
for i in range(1,5):
    if i==2:
        continue

    print("Hello", i)
```

```python
for i in range(1,5):
    if i==2:
        break

    print("Hello", i)
```
