## Practice Problems:

**1. Sum of Numbers :**

**Input:** A positive integer `n`

**Output:** The sum of all numbers from 1 to `n` (inclusive)

**Example:**

```
Input: 5
Output: 15 (1 + 2 + 3 + 4 + 5)
```

**2. Factorial :**

**Input:** A non-negative integer `n`

**Output:** The factorial of `n` (n! = n * (n-1) * ... * 1)

**Example:**

```
Input: 4
Output: 24 (4 * 3 * 2 * 1)
```

**3. Even/Odd Numbers :**

**Input:** A positive integer `n`

**Output:** Print all even numbers from 1 to `n` (inclusive)

**Example:**

```
Input: 8 (to print even numbers)
Output: 2, 4, 6, 8
```

**4. Multiplication Table :**

**Input:** A positive integer `n`

**Output:** Print the multiplication table for numbers 1 to `n`

**Example:**

```
Input: 5
Output:
    1 * 1 = 1
    1 * 2 = 2
    ...
    5 * 5 = 25
```

**5. Palindrome Check :**

**Input:** A string

**Output:** Print "Yes" if the string is a palindrome (reads the same backward as forward), "No" otherwise

**Example:**

```
Input: racecar
Output: Yes
```

**6. Prime Number Check:**

**Input:** A positive integer `n`

**Output:** Print "Yes" if the number is prime (divisible only by 1 and itself), "No" otherwise

**Example:**

```
Input: 11
Output: Yes
```

**7. Fibonacci Sequence :**

**Input:** A non-negative integer `n`

**Output:** Print the first `n` numbers in the Fibonacci sequence (0, 1, 1, 2, 3, 5, ...)

**Example:**

```
Input: 8
Output: 0 1 1 2 3 5 8 13
```

**8. Pattern Printing:**

**Input:** A positive integer `n`

**Output:** Print triangle pattern of numbers or stars using nested loops (e.g., a right triangle of asterisks)

**Example:**

```
Input: 4 (for a right triangle of asterisks)
Output:
    *
    **
    ***
    ****
```

**9. Vowel/Consonant Count:**

**Input:** A string

**Output:** Print the number of vowels and consonants in the string

**Example:**

```
Input: hello world
Output: Vowels: 3, Consonants: 8
```

**10. Reverse a Number:**

**Input:** A positive integer `n`

**Output:** Print the reverse of the number

**Example:**

```
Input: 1234
Output: 4321
```

**11. Find Palindromes:**

**Input:** Two Number `n` and `m`

**Output:** Count of Palindromes between `n` and `m` excluding inputs

**Example:**

```
Input:
n = 100
m = 117
Output: 2

as 101 and 111 are two palindromes between n and m
```
